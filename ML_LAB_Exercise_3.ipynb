```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv('HousingData.csv')


print(" Column Names:\n", df.columns.tolist())
print("\n Shape of the dataset:", df.shape)


print("\n First 5 rows of the dataset:\n", df.head())


print("\n Target Column Description (MEDV):")
print(df["MEDV"].describe())


df = df.dropna()


X = df.drop(columns=["MEDV"])
y = df["MEDV"]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)


model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print("\nMean Squared Error (MSE):", mse)


first_test_input = X_test.iloc[0:1]
predicted_value = model.predict(first_test_input)
print("\n Predicted value for first test example:", predicted_value)
```
